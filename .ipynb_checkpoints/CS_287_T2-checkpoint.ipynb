```python
class TrigramModel(object):
    def __init__(self, alpha1, alpha2, vocab_size, device):
        self.alpha1 = alpha1
        self.alpha2 = alpha2
        self.vocab_size = vocab_size
        self.device = device

    def fit(self, train_iter):
        self.ngram_log_prob = [self._get_ngram_prob(train_iter, n) for n in [1,2,3]]

    def predict(self, batch):
        pass
            
    
    def _find_ngrams(self, input_list, n):
        return zip(*[input_list[i:] for i in range(n)])
    
    def _get_ngram_prob(self, train_iter, n):
        #count ngrams
        count = Counter()
        for batch in iter(train_iter):
            for text in batch.text.transpose('batch','seqlen').values:
                text = text.tolist()
                ngrams = self._find_ngrams(text, n)
                count.update(ngrams)
        
        #calculate probabilities
        total = sum(count.values())
        probs =  {k: (count[k]+1)/(total+self.vocab_size) for k in count} # with laplace smoothing
        return self._convert_probs_to_named_tensor(probs, n)
    
    def _convert_probs_to_named_tensor(self, probs, n):
        #store in numpy matrix
        prob_matrix = np.zeros([self.vocab_size]*n)
        for ngram in probs:
            prob_matrix[ngram] = probs[ngram]
        prob_matrix[prob_matrix == 0] = 1/self.vocab_size
        prob_matrix = np.log(prob_matrix)
        
        #convert to named tensor
        dim_names = ['distribution']
        for i in range(1,n):
            dim_names.append('minus'+str(i))
        dim_names.reverse()
        prob_tensor = NamedTensor(torch.from_numpy(prob_matrix).to(self.device), names = dim_names)
        return prob_tensor
```

The batching is done in a strange way for language modeling. Each element of the batch consists of `bptt_len` words in order. This makes it easy to run recurrent models like RNNs. 

```python
from torchtext.data.iterator import BPTTIterator
from torchtext.data import Batch, Dataset
import math
 

class NamedBpttIterator(BPTTIterator):
    def __iter__(self):
        text = self.dataset[0].text
        TEXT = self.dataset.fields['text']
        TEXT.eos_token = None
        text = text + ([TEXT.pad_token] * int(math.ceil(len(text) / self.batch_size)
                                              * self.batch_size - len(text)))
        data = TEXT.numericalize(
            [text], device=self.device)
        data = (data
            .stack(("seqlen", "batch"), "flat")
            .split("flat", ("batch", "seqlen"), batch=self.batch_size)
            .transpose("seqlen", "batch")
        )

        dataset = Dataset(examples=self.dataset.examples, fields=[
            ('text', TEXT), ('target', TEXT)])
        while True:
            for i in range(0, len(self) * self.bptt_len, self.bptt_len):
                self.iterations += 1
                seq_len = min(self.bptt_len, len(data) - i - 1)
                yield Batch.fromvars(
                    dataset, self.batch_size,
                    text = data.narrow("seqlen", i, seq_len),
                    target = data.narrow("seqlen", i+1, seq_len),
                )
                         
            if not self.repeat:
                return
```

```python
train_iter, val_iter, test_iter = NamedBpttIterator.splits(
    (train, val, test), batch_size=10, device=torch.device("cpu"), bptt_len=32, repeat=False)
```
